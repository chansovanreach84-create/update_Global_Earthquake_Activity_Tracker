- Filtering: Apply multiple boolean conditions using & and | to find earthquakes that are both greater than magnitude 6.0 and occurred in a specific region (e.g., "Japan" or “Texas”).

```python
bool_earthquake = (df['mag'] > 6.0) & ((df['place'] == 'Japan') | (df['place'] == 'Texas'))
earthquake = df[bool_earthquake]
print(earthquake)
```
